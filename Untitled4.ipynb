# **Sales Trend Over Time**

```python
import plotly.express as px
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
monthly_sales = df.groupby(df['ORDERDATE'].dt.to_period("M"))['SALES'].sum().reset_index()
monthly_sales['ORDERDATE'] = monthly_sales['ORDERDATE'].astype(str)

fig = px.line(monthly_sales, x='ORDERDATE', y='SALES',
              title="Monthly Sales Trend", markers=True)
fig.show()
```

# **Top** **Selling** **Categories**

```python
category_sales = df.groupby('PRODUCTLINE')['SALES'].sum().reset_index()

fig = px.bar(category_sales, x='PRODUCTLINE', y='SALES',
             title="Sales by Product Line", text_auto=True)
fig.show()
```

# **Profitability** by **Sub**-**Category**

```python
import plotly.express as px
import pandas as pd

# Example: Group by ProductLine instead of Sub-Category
product_profit = df.groupby('PRODUCTLINE')['SALES'].sum().reset_index()

# Agar Profit column hai to use karo:
# product_profit = df.groupby('PRODUCTLINE')['Profit'].sum().reset_index()

fig = px.bar(product_profit.sort_values(by='SALES', ascending=False),
             x='PRODUCTLINE', y='SALES',  # y='Profit' agar Profit column available ho
             title="Profitability by Product Line",
             text_auto=True)

fig.show()
```

# **Regional** **Sales** **Performance**

```python
region_sales = df.groupby('COUNTRY')['SALES'].sum().reset_index()

fig = px.pie(region_sales, names='COUNTRY', values='SALES',
             title="Sales Share by Country")
fig.show()
```

# **Discount** **vs** **Profit** **Analysis**

```python
# Discount = (MSRP - PRICEEACH)/MSRP
df['Discount'] = (df['MSRP'] - df['PRICEEACH']) / df['MSRP']
```

```python
df['Profit'] = (df['PRICEEACH'] * df['QUANTITYORDERED']) - (df['MSRP'] * df['QUANTITYORDERED'])
```

```python
fig = px.scatter(df, x='Discount', y='Profit',
                 title="Discount vs Profit", trendline="ols")
fig.show()
```

```python
fig = px.scatter(df, x='Discount', y='Profit',
                 color='PRODUCTLINE',
                 title="Discount vs Profit by Product Line",
                 trendline="ols")
fig.show()
```

# **Top 10 Products by Sales**

```python
top_products = df.groupby('PRODUCTCODE')['SALES'].sum().nlargest(10).reset_index()

fig = px.bar(top_products, x='PRODUCTCODE', y='SALES',
             title="Top 10 Products by Sales", text_auto=True)
fig.show()
```

# **Correlation Heatmap (Numerical Features)**

```python
import plotly.figure_factory as ff

corr = df[['SALES', 'QUANTITYORDERED', 'Discount', 'Profit']].corr()

fig = ff.create_annotated_heatmap(
    z=corr.values,
    x=list(corr.columns),
    y=list(corr.index),
    annotation_text=corr.round(2).values,
    colorscale='Viridis'
)
fig.update_layout(title="Correlation Heatmap")
fig.show()
```

# **Profit Margin Calculation** **bold text**

```python
df['Profit_Margin'] = (df['Profit'] / df['SALES']) * 100

fig = px.histogram(df, x='Profit_Margin', nbins=50,
                   title="Distribution of Profit Margin")
fig.show()
```

# **Cumulative Sales (Growth Over Time)**

```python
cumulative_sales = monthly_sales.copy()
cumulative_sales['Cumulative_Sales'] = cumulative_sales['SALES'].cumsum()

fig = px.line(cumulative_sales, x='ORDERDATE', y='Cumulative_Sales',
              title="Cumulative Sales Growth", markers=True)
fig.show()
```

# **Profit vs Quantity Bubble Chart**

```python
fig = px.scatter(df, x='SALES', y='Profit', size='QUANTITYORDERED', color='PRODUCTLINE',
                 title="Profit vs Sales with Quantity Bubble Size")
fig.show()
```
